```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import csv
import pandas as pd
```

```python
marks_csv= pd.read_csv('/Marks.csv')
```

```python
marks_csv
```

```python
subject = marks_csv.Subject
subject
```

```python
marks = marks_csv.Marks
marks
```

```python
plt.plot(subject,marks)
```

```python
xtitle = "Subjects"
ytitle = "Marks Obtained"
```

```python
labels = subject
sizes = marks
```

```python
# Ploting the mark in pie charts

colors = ['lightgray', 'yellow', 'lightblue', 'red', 'green', 'lightpink']   #since we have 6 subjects

explode = (0.1,0,0,0,0,0)   #Specifying the exploding position of pie element

plt.pie(sizes,labels=labels,explode=explode,autopct='%1.2f%%', shadow=True, colors=colors)



plt.legend(title='Index', loc='upper right', bbox_to_anchor=(1,0,0.3,1))

plt.title('Marks Obtained in Pie Chart')

plt.show()
```

```python
# Now let us plot the data in bar graph

fig = plt.figure(figsize=(10,5))    #Height and width setup

plt.bar(labels,sizes,width=0.4,color=colors)

plt.xlabel(xtitle)

plt.ylabel(ytitle)

plt.title('Marks Obtained in Bar-Diagram')

plt.show()
```
